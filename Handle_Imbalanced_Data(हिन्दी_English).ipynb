# INTRODUCTION

Some data sets contain less amount of data for one type of datapoint in a feature, such datasets are called imbalances datasets, so to handle this situation and get the most accurate data, we chose random datapoints from majority datapoints and making them equal to the least amount in order to have a fair analysis and comparasion ground, process is actually quite simple, just choose random values in the dataset equal to minority and concatenate them
<br>
<br>
कुछ डेटासेट्स में **किसी विशेष फ़ीचर के एक प्रकार के डेटा पॉइंट की संख्या बहुत कम होती है**, ऐसे डेटासेट्स को **असंतुलित डेटासेट्स (Imbalanced Datasets)** कहा जाता है।  इस स्थिति को संभालने और **सबसे सटीक परिणाम प्राप्त करने** के लिए, हम **बहुसंख्यक (Majority) डेटा पॉइंट्स में से यादृच्छिक (Random) रूप से कुछ डेटा पॉइंट्स का चयन करते हैं** और उन्हें **अल्पसंख्यक (Minority) डेटा पॉइंट्स की संख्या के बराबर करते हैं**।  **निष्पक्ष विश्लेषण (Fair Analysis) और तुलना (Comparison) का आधार तैयार करना**।  **डेटा असंतुलन (Data Imbalance) के कारण उत्पन्न पूर्वाग्रह (Bias) को कम करना**। **प्रक्रिया (Process) सरल है:**  
**बहुसंख्यक वर्ग (Majority Class) से कुछ डेटा पॉइंट्स यादृच्छिक रूप से चुनें**।  
**उन्हें अल्पसंख्यक वर्ग (Minority Class) की संख्या के बराबर करें**।  
**दोनों को मिलाकर (Concatenate) संतुलित डेटासेट तैयार करें**।  
इस तरह, हमारा मॉडल **असंतुलित डेटा के कारण गलत निष्कर्ष निकालने के बजाय एक संतुलित और निष्पक्ष तरीके से सीखता है**।  
<br>

Below 3 cells are for creating imbalanced datase<br>नीचे दिए गए **तीन सेल असंतुलित (Imbalanced) डेटासेट बनाने के लिए हैं**।  


```python
dictionary = {
    'Gender': ['Male'] * 90 + ['Female'] * 10,
    'Name': ['Person_' + str(i) for i in range(100)],
    'Grade': [85, 78, 92, 45, 76, 88, 90, 55, 79, 95] * 10,
    'Pass_Fail': ['Pass'] * 85 + ['Fail'] * 15
}
```

```python
import pandas as pd
```

```python
dataset = pd.DataFrame(dictionary)
```

```python
dataset.head()
```

As you can see data is imbalanced<br>
जैसा कि आप देख सकते हैं, **डेटा असंतुलित (Imbalanced) है**।  

```python
dataset['Pass_Fail'].value_counts()
```

We saperated the dataset into two parts containing each group<br>हमने डेटासेट को **दो भागों में विभाजित किया**, जहाँ प्रत्येक भाग में **एक-एक समूह के डेटा पॉइंट्स** शामिल हैं।  

```python
pass_ = dataset[dataset['Pass_Fail'] == 'Pass']
fail_ = dataset[dataset['Pass_Fail'] == 'Fail']
```

We chose datasets/dataset with majority and randomly chose datapoints in number equal to minority<br>हमने **बहुसंख्यक (Majority) डेटा सेट** को चुना और उसमें से **यादृच्छिक रूप से उतने ही डेटा पॉइंट्स का चयन किया**, जितने **अल्पसंख्यक (Minority) डेटा सेट में मौजूद थे**।  

```python
passed_student_sample = pass_.sample(n=15)
```

```python
df_new = pd.concat([passed_student_sample, fail_], axis=0)
```

Now we will work with this dataset
<br>अब हम इस **संतुलित (Balanced) डेटासेट** पर कार्य करेंगे।
